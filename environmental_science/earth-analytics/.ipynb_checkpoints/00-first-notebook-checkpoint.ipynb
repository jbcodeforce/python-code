```python
import matplotlib.pyplot as plt
import pandas as pd

# Create data
boulder_precip = pd.DataFrame(columns=["date", "precip"], 
                              data=[
                                  ["2013-09-09", 0.1], ["2013-09-10", 1.0], ["2013-09-11", 2.3], ["2013-09-12", 9.8], ["2013-09-13", 1.9],
                                  ["2013-09-14", 0.01], ["2013-09-15", 1.4], ["2013-09-16", 0.4]])      
# Create plot
fig, ax = plt.subplots()
ax.bar(boulder_precip['date'].values, boulder_precip['precip'].values)
ax.set(title="Daily Precipitation (inches)\nBoulder, Colorado 2013", 
       xlabel="Date", ylabel="Precipitation (Inches)")
plt.setp(ax.get_xticklabels(), rotation=45);
```

# Pandas Data frame
Pandas dataframes are composed of rows and columns that can have header names, and the columns in pandas dataframes can be different types. Each column can have a label name. 
Each row has an index within a range of values beginning at [0].
You can plot pandas dataframe using matplotlib or using the pandas .plot() method which wraps around matplotlib.


```python
import earthpy as et
import os
# URL for .csv with avg monthly precip data in Colorado
avg_monthly_precip_url = "https://ndownloader.figshare.com/files/12710618"

current_directory = os.getcwd()
et.data.path = current_directory
# Download file from URL
data_path=et.data.get_data(url=avg_monthly_precip_url)
print(data_path)
# Set working directory to earth-analytics/data
os.chdir(os.path.join(current_directory))
```

```python
fname = os.path.join("earthpy-downloads",
                     "avg-precip-months-seasons.csv")

avg_monthly_precip = pd.read_csv(fname)

avg_monthly_precip
```

```python
fig, ax = plt.subplots()
ax.bar(avg_monthly_precip['months'].values, avg_monthly_precip['precip'].values)
ax.set(title="Monthly Precipitation (inches)", 
       xlabel="Date", ylabel="Precipitation (Inches)")
```

```python
# Describe Contents of Pandas Dataframes
avg_monthly_precip.info()
```

```python
# Number of rows and columns
avg_monthly_precip.shape
```

```python
# Summary stats of all numeric columns
avg_monthly_precip.describe()
```

```python
# Median of all numeric columns
avg_monthly_precip['precip'].median()
```

```python
avg_monthly_precip.sort_values(by='precip', ascending=False)
```

```python
# Group data by seasons and summarize precip
precip_by_season=avg_monthly_precip.groupby(["seasons"])[["precip"]].describe()
precip_by_season
```

```python
# Drop a level so there is only one index
precip_by_season.columns = precip_by_season.columns.droplevel(0)
# Plot the data
f, ax = plt.subplots()

ax.bar(precip_by_season.index,
        precip_by_season["mean"],
        color="purple")

ax.set(title="Bar Plot of Seasonal Monthly Precipitation in mm")
plt.show()
```
